## **TRIMMING SECTION** 

Video trim section

```python
################
# TRIMMING GUI # 
################
# START TIME 
lbl_starttime=widgets.Label('Start Time (in secs)')
display(lbl_starttime)
txt_starttime=widgets.BoundedIntText(min=0)
display(txt_starttime)

# END TIME
lbl_endtime=widgets.Label('End Time (in secs)')
display(lbl_endtime)
txt_endtime=widgets.BoundedIntText(min=0)
display(txt_endtime)

# VIDEO NAME 
lbl_fname=widgets.Label('File Name [❌] NAME MUST BE UNIQUE ')
display(lbl_fname)
txt_fname=widgets.Text()
display(txt_fname)

# BUTTON 
btn=widgets.Button(description="ADD FILE")
display(btn)

# btn.on_click()
```

```python
from moviepy.video.io.VideoFileClip import VideoFileClip

#Change the path according to your folder name. 
videopath = fc.value
input_video_path = videopath
output_video_path = '/content/drive/MyDrive/pipline/Video/{}.mp4'.format(txt_fname) 

##############
# TRIM LOGIC #
##############
with VideoFileClip(input_video_path) as video:
    new = video.subclip(0, 60)
    new.write_videofile(output_video_path, audio_codec='aac')
```
